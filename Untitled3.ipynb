```python
iron = int(input())
bat = int(input())
total_suits = int(input())
total_batt = int(input())

# 特例：iron == bat
if iron == bat:
    # 所有裝備必須都是同一種，且 x>0 y>0 → total_suits >= 2
    if total_batt == iron * total_suits and total_suits >= 2:
        print("True")
    else:
        print("False")
    exit()

# 一般情況，不使用浮點數（避免浮點誤差與 RE）
diff = iron - bat
rhs = total_batt - bat * total_suits

# 整除檢查
if rhs % diff != 0:
    print("False")
    exit()

x = rhs // diff
y = total_suits - x

if x > 0 and y > 0:
    print("True")
else:
    print("False")
```
